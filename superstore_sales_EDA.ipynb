```python
import seaborn as sns
import matplotlib.pyplot as plt
corr = superstore[["Sales", "Profit", "Quantity", "Discount", "Shipping Days"]].corr()
```

```python
plt.figure(figsize= (6, 4))
sns.heatmap(
    corr,
    annot=True,
    cmap = "coolwarm",
    fmt = ".2f"
)
plt.title("Correlation heatmap")
plt.show()
```

THE SALES AND PROFIT SHARE A POSITIVE CORRELATION OF **0.48**.
**THE DISCOUNT DRAIN** : DISCOUNT SHOWS STARK NEGATIVE CORRELATION WITH PROFIT OF **-0.22** AND ZERO CORRELATION WITH SALES (**-0.03**). **WE ARE SACRIFICING MARGIN WITHOUT EVEN MAKING ANY TOP-LINE REVENUE LIFT.**

SHIPPING DAYS SHOW A NEAR **ZERO CORRELATION** WITH ALL THE FINANCIAL AND VOLUME METRICS, **INDICATING DELIEVERY SPEED IS NOT A DRIVER OF PURCHASING BEHAVIOR.**
